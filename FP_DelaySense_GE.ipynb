```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-data-FP'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'FP_GX'
path_to_data = '/content/Train.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

# Expectation Suites

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-delivery-data'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

```python
# Expectation 1 : All columns can not contain missing values
for column in validator.active_batch.columns():
    result = validator.expect_column_values_to_not_be_null(column)
    print(f"Column: {column}")
    print(result)
```

```python
# Expectation 2 : Column `ID` must be unique

validator.expect_column_values_to_be_unique('ID')
```

```python
# Expectation 3 : Column `gender` must contain M and F

validator.expect_column_values_to_be_in_set('Gender', ['M', 'F'])
```

```python
# Expectation 4 : Column `Mode_of_Shipment` must contain Flight, Ship, Road.
validator.expect_column_values_to_be_in_set('Mode_of_Shipment', ['Flight', 'Ship', 'Road'])
```

```python
# Expectation 5 : Column `Warehouse_block` must contain A, B, C, D, and F.

validator.expect_column_values_to_be_in_set('Warehouse_block', ['A', 'B', 'C', 'D', 'F'])
```

```python
# Expectation 6 : Column `Product_importance` must contain low, medium, high.

validator.expect_column_values_to_be_in_set('Product_importance', ['low', 'medium', 'high'])
```

```python
# Expectation 7 : Column `Reached.on.Time_Y.N` must contain 0 and 1.

validator.expect_column_values_to_be_in_set('Reached.on.Time_Y.N', [0, 1])
```

```python
# Expectation 8 : Column `numeric` must in form of integer or float

columns_check = ['ID', 'Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product',
                 'Prior_purchases', 'Discount_offered', 'Weight_in_gms', 'Reached.on.Time_Y.N']
types = ['int64']

for column in columns_check:
    result = validator.expect_column_values_to_be_in_type_list(column, types)
    print(f"Column: {column}")
    print(result)
```

```python
# Expectation 8 : The maximum value of column `Customer_rating ` must be `$ 0.5`

validator.expect_column_max_to_be_between('Customer_rating', 1, 5)
```

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```

# Checkpoint

```python
# Create a checkpoint

checkpoint_FP = context.add_or_update_checkpoint(
    name = 'checkpoint_FP',
    validator = validator,
)
```

```python
# Run a checkpoint

checkpoint_result = checkpoint_FP.run()
```
